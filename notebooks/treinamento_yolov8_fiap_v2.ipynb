## Treinamento Inicial do Modelo

Nesta etapa é realizado um treinamento curto de 5 épocas utilizando o modelo **YOLOv8n**. O objetivo é validar a configuração do ambiente, verificar se o dataset foi carregado corretamente e confirmar que o treinamento ocorre sem erros antes da execução do treinamento completo.

```python
!ls /content
```

```python
from google.colab import files

uploaded = files.upload()
```

```python
!ls /content
```

```python
from pathlib import Path

assert Path("/content/best.pt").exists(), "best.pt não encontrado"
assert Path("/content/dataset_v4/dataset_colab.yaml").exists(), "YAML não encontrado"

print("best.pt encontrado")
print("dataset_v4 encontrado")
```

```python
from ultralytics import YOLO

# Carrega o modelo treinado anteriormente
model = YOLO("/content/best.pt")

results = model.train(
    data="/content/dataset_v4/dataset_colab.yaml",
    epochs=25,
    imgsz=640,
    batch=16,
    workers=2,
    cache=False,
    device=0,
    patience=8,
    project="/content/runs",
    name="fiap_finetuning_v4",
    exist_ok=True
)
```

## Resultado do Fine-Tuning

O processo de fine-tuning foi concluído com sucesso após 25 épocas, utilizando como ponto de partida o modelo previamente treinado (best.pt) e o novo dataset_v4, composto por aproximadamente 1.200 imagens sintéticas contendo componentes genéricos, AWS e Azure.

As métricas obtidas demonstraram excelente desempenho, alcançando Precision de 99,9%, Recall de 99,8%, mAP50 de 99,5% e mAP50-95 de 97,7% no conjunto de validação. Esses resultados indicam que o modelo conseguiu aprender corretamente as novas variações de componentes e manter elevado desempenho na detecção das quinze classes utilizadas no projeto.

Com a conclusão do fine-tuning, o modelo está apto para a etapa de validação em diagramas reais, permitindo avaliar sua capacidade de generalização em arquiteturas AWS e Azure antes da integração definitiva ao pipeline da aplicação.
